```python
# imports
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.linear_model import LinearRegression 
from matplotlib import pyplot as plt


from pathlib import Path
```

```python
file_path = Path("clean_selected_covid_data2.csv")
df = pd.read_csv(file_path)
df.head(5)
```

```python
X = df.new_vaccinations_smoothed.values.reshape(-1, 1)
y = df.icu_patients_per_million
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
print(y_pred.shape)
plt.scatter(X, y)
plt.xlabel('New Vaccinations')
plt.ylabel('ICU Patients')
plt.title('Figure. 1: Vaccinations vs ICU Patients')
plt.plot(X, y_pred, color='red')
plt.show()
```

```python
X = df.new_vaccinations_smoothed.values.reshape(-1, 1)
y = df.new_deaths_smoothed
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
print(y_pred.shape)
plt.scatter(X, y)
plt.xlabel('New Vaccinations')
plt.ylabel('New Deaths')
plt.title('Figure. 2: Vaccinations vs Deaths')
plt.plot(X, y_pred, color='red')
plt.show()
```

```python
X = df.new_cases_smoothed.values.reshape(-1, 1)
y = df.new_deaths_smoothed
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
print(y_pred.shape)
plt.scatter(X, y)
plt.xlabel('New Cases')
plt.ylabel('New Deaths')
plt.title('Figure. 3: Positive Cases vs Deaths')
plt.plot(X, y_pred, color='red')
plt.show()
```
